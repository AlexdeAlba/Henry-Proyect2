Guardamos en un archivo los datos, los abrimos, los cargamos en un DF e  imprimimos el DF y su tamaño y revisamos los tipos de datos que contiene.

```python
import os
import pandas as pd

directorio = '.\data'
files = []
name_files = []

# Obtener la lista de archivos en el directorio
archivos = os.listdir(directorio)
# Filtrar solo los archivos con extensión .csv y mostrar la ruta completa
for archivo in archivos:
    if archivo.endswith('.csv'):
        file = os.path.join(directorio, archivo)
        name_files.append(archivo)
        files.append(file)
# Cargar cada archivo CSV en un DataFrame y agregarlo a la lista dfs
dfs = []

print("Archivos incluidos en el analisis:")
for i, name in enumerate(name_files):
    print(f'Archivo {i+1}: {name}')

for file in files:
    df = pd.read_csv(file)
    dfs.append(df)

#Imprime los DF
columnas = []
for count, dfi in enumerate(dfs):
    print(f'Dataframe num {count+1} : {name_files[count]}  Tamaño: {len(dfi)}')
    display(dfi.head(3))
```

## Buscando valores nulos

Como primer paso se eliminan todas las columnas y las filas que contengan puros nulos

```python
#Limpia las columnas y filas con valores nulos
def limpiar_df(df):
    # Eliminar columnas con todos sus valores nulos
    df = df.dropna(axis=1, how='all')
    # Eliminar registros con todas las columnas nulas
    df = df.dropna(how='all')
    return df  # Devuelve el DataFrame limpio

# Itera a través de la lista de DataFrames y 
# actualiza cada DataFrame con el resultado de la limpieza
for i, dfi in enumerate(dfs):
    dfs[i] = limpiar_df(dfi)
```

```python
df_nulos = []
for count, dfi in enumerate(dfs):
    if dfi.isnull().values.any():
        df_nulos.append(dfi.copy())
        display(dfi.head(3))

print("******************************************")
print("******************************************")
print("******************************************")
print("")
    
```

Graficamos los nulos de cada columna

```python
import matplotlib.pyplot as plt
import missingno

# Funcion que grafica los nulos de un df
def plot_nulls(df):
    # Obtener la cantidad de valores nulos por columna
    valores_nulos = df.isna().sum()

    # Filtrar las columnas con valores nulos
    columnas_con_nulos = valores_nulos[valores_nulos > 0].index

    # Generar una lista de colores basada en la cantidad de columnas con nulos
    colores = plt.cm.tab20(range(len(columnas_con_nulos)))
    #colores = lista_de_colores(df)

    #Imprime las barras de los nulos
    missingno.bar(df[columnas_con_nulos], fontsize=8, color=colores, figsize=(10, 3))
    plt.title("Valores no nulos", fontsize=18)
    plt.show()

import seaborn as sns

def lista_de_colores(df):
    num_columnas = len(df.columns)
    colores = sns.color_palette("hls", num_columnas)
    print("Numero de colores asignados: ",len(colores))
    return colores    
```

```python
for count, dfi in enumerate(df_nulos):
    #display(dfi.head(0))
    valores_nan = dfi.isna()
    nanXcolumna = pd.DataFrame({'Valores':valores_nan.sum(), 'Tamaño' : len(dfi)})
    nanXcolumna.transpose()
    print("Dataframe: ",name_files[count])
    display(nanXcolumna)
    # Grafico los nulos
    plot_nulls(dfi)
    print("******************************************")
    print("******************************************")
    print("******************************************")
    print("")
    
```

# Dataset externos

El dataset de población y de ubicacion fue extraido de [argentina.gob.ar](https://www.argentina.gob.ar/pais/poblacion)

```python
import pandas as pd
import matplotlib.pyplot as plt


# ****   Censo ********
# Lee el df (Read the DataFrame)
file1 = 'dataClean\\censo_poblacion.csv'
df_censo = pd.read_csv(file1)
# Aplicar el formato a la columna provincia_nombre
df_censo['provincia_nombre'] = df_censo['provincia_nombre'].str.title()
# Agrupar por provincia_id y provincia_nombre y sumar la cantidad
df_censo = df_censo.groupby(['provincia_id', 'provincia_nombre'])['cantidad'].sum().reset_index()
# Renombrar la columna 'cantidad' a 'poblacion_total'
df_censo = df_censo.rename(columns={'cantidad': 'poblacion_total'})

#****  Ubicacion *******
file2 = 'dataClean\\provincias_ubicacion.csv'
df_ubicacion = pd.read_csv(file2,encoding='latin-1')

# Union de los df por provincia_nombre y Provincia
df_unido = pd.merge(df_censo, df_ubicacion, left_on='provincia_nombre', right_on='Provincia')

# Eliminar la columna duplicada (Provincia) si es necesario
df_unido = df_unido.drop('Provincia', axis=1)


display(df_censo.head(2))
display(df_ubicacion.head(2))
display(df_unido.head(3))

# Guardar el DataFrame como un archivo CSV
df_unido.to_csv('dataKPIs//provincias.csv', index=False)
```
